###Desafio 10 - Na tupla de retorno do forecast, estudar na documentação o que são cada um dos elementos de retorno

https://www.statsmodels.org/devel/generated/statsmodels.tsa.arima_model.ARIMAResults.forecast.html#statsmodels.tsa.arima_model.ARIMAResults.forecast

```
from statsmodels.tsa.arima_model import ARIMA

dados_treino = dados_st.query("Ano != 2019")
dados_teste = dados_st.query("Ano == 2019").copy()

modelo_arima = ARIMA(dados_treino["Casos"].values.astype("float32"), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
```

```
forecast, stderr, conf_int = modelo_arima_treinado.forecast(12)
```

```
forecast
```

```
forecast + stderr
```

```
forecast - stderr
```

```
conf_int
```

```
dados_teste["forecast"] = forecast
dados_teste["forecast + stderr"] = forecast + stderr
dados_teste["forecast - stderr"] = forecast - stderr
dados_teste["conf_int0"] = [valor[0] for valor in conf_int]
dados_teste["conf_int1"] = [valor[1] for valor in conf_int]

plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Data", y="forecast", data=dados_teste, label='previsão')
ax.fill_between(dados_teste['Data'], dados_teste["forecast - stderr"], dados_teste["forecast + stderr"], color='r', alpha=.1, label='Erro padrão da previsão')
ax.fill_between(dados_teste['Data'], dados_teste["conf_int0"], dados_teste["conf_int1"], color='g', alpha=.1, label='Intervalo de confiança da previsão')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title('Entendendo o retorno do método fit do modelo ARIMA')
plt.show()
```
