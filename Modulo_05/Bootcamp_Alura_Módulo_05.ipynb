### Desafio 01 - Tipos de aprendizados

Pesquisar aprendizado supervisionado, não supervisionado, por reforço e mais

Aprendizado **supervisionado**

- treinados por meio de exemplos rotulados
-  utiliza como entrada um conjunto prévio de dados coletados do mundo real e utilizado para "treinamento";
- todas as entradas e saídas são conhecidas;
- ex.: transaçõa via cartão de crédito (fraude ou não), emails (SPAM ou não)


---


Aprendizado **não supervisionado**

- o algoritmo não recebe um conjunto de dados prévio;
- objetivo: explorar os dados e encontrar algum padrão dentro deles;
- o sistema é alimentado apenas com dados de entrada;


---


Aprendizado **por reforço**

- pode ser utilizado para as mesmas aplicações que o aprendizado supervisionado;
- manipula tanto dados rotulados quanto não-rotulados para treinamento;
- pode ser empregado com métodos como classificação, regressão e previsão;
- útil quando o custo associado à rotulação é muito alto;

---


Aprendizado **por reforço**

- o sistema requalifica suas regras de avaliação com base em feedbacks observados do mundo;
-  o algoritmo descobre através de testes do tipo 'tentativa e erro' quais ações rendem as maiores recompensas;
- geralmente utilizado em robótica, jogos e navegação;

### Desafio 02 - Subtipos de aprendizados

Pesquisar sobre regressão, classificação e etc.

Dentro do **aprendizado supervisionado**, temos:

---

**Classificação**

- processo de tomar algum tipo de entrada e atribuir um rótulo a ela;
- usados ​​geralmente quando as previsões são de natureza distinta, ou seja, binárias;

---

**Regressão**

- usada quando o valor que está sendo previsto difere de algo binário e siga um espectro contínuo;
- ex.: quanto custa, quantos existem, etc;


Dentro do **aprendizado NÃO supervisionado**, temos:

---

**Clustering (ou agrupamento)**

- agrupa os dados em clusters (grupos) baseado em suas características/variáveis;
- geralmente é utilizado para uma melhor visualização dos dados, para estudar padrões de comportamento e parar detectar outliers.

### Desafio 05 - Tamanho dos dados de teste e treino

Decida quanto você vai escolher para treino e quanto para teste? Quanto as pessoas utilizam e qual o motivo? Olhar a documentação

A resposta é que, como regra prática, você precisa de aproximadamente 10 vezes mais exemplos do que os *graus de liberdade* em seu modelo.

No caso de um modelo linear, os graus de liberdade são essencialmente iguais à dimensionalidade dos dados (um número de colunas). Descobrimos que pensar em termos de dimensionalidade versus número de exemplos é um atalho conveniente.

Quanto mais poderoso o modelo, mais ele está sujeito a overfitting e, portanto, mais exemplos você precisa. E, claro, a maneira de controlar isso é por meio da validação.

```
url = 'https://github.com/marcelagomescorrea/bootcamp_alura/blob/main/Modulo_05/datasets/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx?raw=true'
dados = pd.read_excel(url)
dados.head()
```

```
features_continuas = dados.iloc[:, 13:-2].columns
features_continuas
```

```
def preenche_tabela(dados):
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby('PATIENT_VISIT_IDENTIFIER',
                                       as_index=False)[features_continuas_colunas]\
                                       .fillna(method='bfill')\
                                       .fillna(method='ffill')
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]

    dados_finais = pd.concat([features_categoricas, features_continuas, saida],
                             ignore_index=True, axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

```
dados_limpos = preenche_tabela(dados)
dados_limpos.head()
```

```
dados_limpos.describe()
```

```
dados_limpos = dados_limpos.dropna()
```

```
a_remover = dados_limpos.query('WINDOW == "0-2" and ICU==1')['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query('PATIENT_VISIT_IDENTIFIER not in @a_remover')
dados_limpos.describe()
```

```
def prepare_window(rows):
    if(np.any(rows['ICU'])):
        rows.loc[rows['WINDOW'] == '0-2', 'ICU'] = 1

    return rows.loc[rows['WINDOW'] == '0-2']
```

```
dados_limpos = dados_limpos.groupby('PATIENT_VISIT_IDENTIFIER').apply(prepare_window)
dados_limpos.head()
```

```
y = dados_limpos['ICU']
X = dados_limpos[x_columns].drop(['ICU'], axis=1)
```

```
np.random.seed(74246)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```
modelo_dummy = DummyClassifier()
modelo_dummy.fit(X_train, y_train)

y_prediction_dummy = modelo_dummy.predict(X_test)
accuracy_score(y_test, y_prediction_dummy) * 100
```

```
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

y_prediction = modelo.predict(X_test)
accuracy_score(y_test, y_prediction) * 100
```
